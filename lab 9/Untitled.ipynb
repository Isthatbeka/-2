```python
# scatter plot  BMI  Glucose,    
#(25 /**2  140 mg/dL, ).  -      Outcome.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
diabetes = pd.read_csv("diabetes.csv")
colors = [ 'royalblue', 'sienna']
lc = ['magenta', 'yellow', 'lime']
lines = diabetes.shape[0]
x = []
numbers = [25, 140]
type = ("BMI", "Glucose")
plt.figure(figsize = (35,15))

for i in range(lines):
    x.append(i+1)
    
for i in range (len(type)):    
    plt.scatter(x, diabetes[type[i]], color = colors[i],s = 100)
    
    plt.legend(type,loc = 'upper left',fontsize=15)
for i in range (len(type)):  
    plt.hlines(numbers[i], 0, 1000,colors = lc[i])
    #plt.legend()
plt.show
```
