```python
import matplotlib.pyplot as plt
import pandas as pd

students = pd.read_csv("StudentsPerformance.csv")

a = students[students["parental level of education"] == "bachelor's degree"] 
b = students[students["parental level of education"] == "some high school"]
c = students[students["parental level of education"] == "some college"] 
d = students[students["parental level of education"] == "master's degree"]
e = students[students["parental level of education"] == "high school"] 
f = students[students["parental level of education"] == "associate's degree"] 


e_l_types = [a,b,c,d,e,f]
e_l_types1 = ["associate's degree", "bachelor's degree", "college", "high school", "some high school", "masters degree"]
fig, sp = plt.subplots(6, figsize=(15, 15))

for i in range(len(e_l_types1)):
    sp[i].set_xlabel('Reading score')
    sp[i].set_ylabel('Math score')
    sp[i].scatter(e_l_types[i]["reading score"], e_l_types[i]["math score"])
    sp[i].legend(title = e_l_types1[i])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

students = pd.read_csv("StudentsPerformance.csv")

types=["reading score", "math score", "writing score"]
a = students[students["parental level of education"] == "bachelor's degree"] 
b = students[students["parental level of education"] == "some high school"]
c = students[students["parental level of education"] == "some college"] 
d = students[students["parental level of education"] == "master's degree"]
e = students[students["parental level of education"] == "high school"] 
f = students[students["parental level of education"] == "associate's degree"]
exams = ["math score", "reading score", "writing score"]
e_l_types =  ["bachelors\ndegree", 'some\nhigh\nschool', "some\ncollege", 'masters\ndegree', 'high\nschool', 'associates\ndegree']

for i in range (len(types)):
    plt.figure(figsize=(15, 5))
    plt.boxplot((a[exams[i]], b[exams[i]], c[exams[i]], d[exams[i]], e[exams[i]], f[exams[i]]), labels=e_l_types)
    plt.title('Parental level of education,\n'+types[i])
    plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

mice = pd.read_csv('miceprotein.csv', na_values='?').dropna()

mice_classes=[*mice['class'].unique()]
print(mice_classes)

level_of_exp=[]
for i in range (len(mice_classes)):
    level_of_exp.append(mice[mice['class']==mice_classes[i]])
plt.figure(figsize=(15, 5))
for i in range (len(level_of_exp)):
    sp=plt.subplot(1,len(level_of_exp), i+1)
    sp.boxplot(level_of_exp[i]['SOD1_N'].dropna(), labels=[mice_classes[i]])
    plt.title('SOD1_N,\n'+mice_classes[i])    
plt.figure(figsize=(15, 5))
for i in range (len(level_of_exp)):
    sp=plt.subplot(1,len(level_of_exp), i+1)
    sp.boxplot(level_of_exp[i]['DYRK1A_N'].dropna(), labels=[mice_classes[i]])
    plt.title('DYRK1A_N,\n'+mice_classes[i])

plt.show()
```
